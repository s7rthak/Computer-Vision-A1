```python
import os
import cv2
import argparse
import numpy as np

BASELINE_FRAMES = 1700
ILLUMINATION_FRAMES = 301
JITTER_FRAMES = 1150
DYNAMIC_FRAMES = 1189
PTZ_FRAMES = 1130

backSub = cv2.createBackgroundSubtractorMOG2(varThreshold=30, detectShadows=False)
kernel = np.ones((5,5),np.uint8)

for i in range(1, BASELINE_FRAMES+1):
    frame_name = "in" + str(i).zfill(6) + ".jpg"

    frame = cv2.imread("COL780-A1-Data/baseline/input/" + frame_name)

    bilateral = cv2.bilateralFilter(frame, 15, 75, 75)
    fgMask = backSub.apply(bilateral)
    closing = cv2.morphologyEx(fgMask, cv2.MORPH_CLOSE, kernel)
    opening = cv2.morphologyEx(closing, cv2.MORPH_OPEN, kernel)

    # cv2.imshow('Frame', frame)
    cv2.imshow('FG Mask', opening)

    keyboard = cv2.waitKey(30)
    if keyboard == 'q' or keyboard == 27:
        break
```
