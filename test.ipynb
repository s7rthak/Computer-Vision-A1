```python
import random 
import subsense

class Subsense_background:
    def __init__(self,shape,d_init,v_init,r_init,t_init,r_color_0,r_lbsp_0,n_min,learn_rate):
        self.h, self.w, self.c = shape
        self.background_model = []
        self.prev_map = np.full((self.h,self.w),0)
        for i in range(0,50):
            x1 = random.randint(0,self.h-1)
            y1 = random.randint(0,self.w-1)
            self.background_model.append((x1,y1))
        
        self.d_min = np.full((self.h,self.w),d_init)
        self.v = np.full((self.h,self.w),v_init)
        self.r = np.full((self.h,self.w),r_init)
        self.t = np.full((self.h,self.w),t_init)
        self.r_color_0 = r_color_0
        self.r_lbsp_0 = r_lbsp_0
        self.n_min = n_min
        self.learn_rate = learn_rate
        
    def apply(self, img):
        r_color = self.r * self.r_color_0
        r_lbsp = np.power( np.full((self.h,self.w),2), self.r) + self.r_lbsp_0
        self.curr_segmentation = np.full((self.h,self.w),1)
        self.lbsp,self.intensity = subsense.compute_lbsp(img, 0.3)
        self.distance_tot = np.full((self.h,self.w),0)
        
        for i in range(0,self.h):
            for j in range(0,self.w):
                pixel_lbsp = self.lbsp[i,j]
                pixel_intensity = self.intensity[i,j]
                no_matches = 0
                sum_store = 0
                for k in range(0,len(self.background_model)):
                    temp_pixel_lbsp = self.lbsp[self.background_model[k][0],self.background_model[k][1]]
                    temp_pixel_intensity = self.intensity[self.background_model[k][0],self.background_model[k][1]]
                    dist_lbsp = subsense.hamming_distance(pixel_lbsp,temp_pixel_lbsp)
                    dist_intensity = abs(temp_pixel_intensity - pixel_intensity)
                    sum_store = sum_store + dist_lbsp
                    sum_store = sum_store + dist_intensity
                    if dist_lbsp > r_lbsp[i,j]:
                        continue
                    if dist_intensity > r_color[i,j]:
                        continue
                    no_matches = no_matches + 1
                self.distance_tot[i,j] = sum_store
                if no_matches >= self.n_min:
                    self.curr_segmentation[i,j] = 0
                store = random.uniform(0, 1)
                if store <= self.t[i,j]:
                    idx = random.randint(0,len(self.background_model)-1)
                    self.background_model[idx] = (i,j)
                
                
        #Now for updates
        mean = np.mean(self.distance_tot.flatten())
        variance = np.std(self.distance_tot.flatten())
        self.distance_tot = (self.distance_tot - mean)/ variance
        self.d_min = (1- self.learn_rate)* self.d_min + self.learn_rate * self.distance_tot
        
        self.xor_store = self.curr_segmentation ^ self.prev_map
        self.prev_map = self.curr_segmentation
        
        for i in range(0,self.h):
            for j in range(0,self.w):
                if self.xor_store[i,j] == 1 :
                    self.v[i,j] = self.v[i,j] + 1
                else:
                    self.v[i,j] = self.v[i,j] - 0.1

        
        for i in range(0,self.h):
            for j in range(0,self.w):
                if self.r[i,j] >= (1 + self.d_min[i,j]*2) ** 2 :
                    self.r[i,j] = self.r[i,j] + self.v[i,j]
                else:
                    self.r[i,j] = self.r[i,j] - (1/self.v[i,j])
                    
        for i in range(0,self.h):
            for j in range(0,self.w):
                if self.curr_segmentation[i,j] == 1:
                    self.t[i,j] = self.t[i,j] + 1/(self.v[i,j]*self.d_min[i,j])
                else:
                    self.t[i,j] = self.t[i,j] - (self.v[i,j]/self.d_min[i,j])
        return self.curr_segmentation
        
        

        
```
